```python
import numpy as np 
import pandas as pd 

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score,recall_score
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from imblearn.over_sampling import RandomOverSampler
import warnings
warnings.filterwarnings( 'ignore' )


#모든 값 -1로하면 46.815

df = pd.read_csv('uci-secom.csv')
df = df.drop(columns = ['Time'], axis = 1)
data = df.replace(np.NaN, 0)

x = data.drop(columns = ['Pass/Fail'], axis = 1)
y = data['Pass/Fail']
y=np.array(y).ravel()

ros = RandomOverSampler()

oversampled_data, oversampled_label = ros.fit_resample(x, y)
oversampled_data = pd.DataFrame(oversampled_data)
oversampled_label = pd.DataFrame(oversampled_label)

x_train, x_test, y_train, y_test = train_test_split(oversampled_data,oversampled_label, test_size = 0.2, random_state = 0)

#정규화
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
#모델 학습 코드

#gridsearchCV
parameters = {'C': [0.001, 0.01, 0.1], 'gamma' : [0.001, 0.01, 0.1, 1]}
SVM = SVC(kernel='rbf')
SVM = GridSearchCV(SVM, param_grid=parameters, scoring = 'accuracy', cv = 5)
SVM.fit(x_train, y_train) 
print('train :',SVM.score(x_train, y_train))
print('test:',SVM.score(x_test, y_test))
print('best_param:',SVM.best_params_)

#베스트값찾기
best_SVM_CV = SVM.best_estimator_

#학습데이터 최종 출력
model_predition=best_SVM_CV.predict(x_test)
print("accuracy score: {}".format(accuracy_score(y_test, model_predition)))
print("Recall score: {}".format(recall_score(y_test, model_predition)))
```

```python
dataframe = pd.read_csv("manufacture_test_feature.csv")
dataframe = dataframe.replace(np.NaN, 0)

test_x = dataframe.drop(columns = ['Time'], axis = 1)
test_x = sc.transform(test_x) 

result=best_SVM_CV.predict(test_x)

model_predition=pd.DataFrame(result,columns=['Pass/Fail'])
model_predition
```
