### 3) 알고리즘 연습

- **`알고리즘(algorithm)`**은 수학과 컴퓨터 과학, 언어학 또는 관련 분야에서 `어떠한 문제를 해결하기 위해 정해진 일련의 절차나 방법`을 공식화한 형태로 표현한 것, 계산을 실행하기 위한 단계적 절차를 의미한다. (wikipedia)<p>
- **`알고리즘`**은 연산, 데이터 진행 또는 자동화된 추론을 수행한다. 산법(算法), 셈법, 계산 절차 등으로 번역되기도 한다.<p>
- 다음은 좋은 알고리즘의 특징이다.
    - 정밀성 : 변하지 않는 명확한 작업 단계를 가져야 한다.
    - 유일성 : 각 단계마다 명확한 다음 단계를 가져야 한다.
    - 타당성 : 구현할 수 있고 실용적이어야 한다.
    - 입력 : 정의된 입력을 받아들일 수 있어야 한다.
    - 출력 : 답으로 출력을 내보낼 수 있어야 한다.
    - 유한성 : 특정 수의 작업 이후에 정지해야 한다.
    - 일반성 : 정의된 입력들에 일반적으로 적용할 수 있어야 한다.

#### 1) 절댓값 구하기

- (방법1) a가 0보다 큰거나 같은지를 판단하여 a가 0보다 크거나 같다면 a를 반환하고, a가 0보다 작다면 -a를 반환

```python
def abs_num(x):
    if x>0:
        return x
    else:
        return -x
```

```python
print(abs_num(1))
print(abs_num(-2))
print(abs_num(0))
```

- (방법2) 제곱근을 활용하여 a에 제곱을 취해주고 다시 제곱근을 반환

```python
import math

def abs_sqrt(x):
    res = x**2
    return math.sqrt(res)
```

```python
print(abs_sqrt(1))
print(abs_sqrt(-2))
print(abs_sqrt(0))
```

#### 2) 1 ~ n까지 합을 구하기

- (방법1) 1부터 n까지 순차적으로 값을 더해 결과를 반환

```python
def sum_num(x):
    res = 0
    for i in range(1,x+1):
        res += i
    return res
```

```python
print(sum_num(10))
print(sum_num(100))
print(sum_num(1000))
# 계산 복잡도 n
```

- (방법2) 1부터 n까지의 합을 구하는 공식을 활용하여 결과를 반환

```python
def sum_equ(x):
    return x*(x+1)//2
```

```python
print(sum_equ(10))
print(sum_equ(100))
print(sum_equ(1000))
# 계산 복잡도 1
```

#### * 계산 복잡도( Big-O )
- **`계산 복잡도 이론(Computational complexity theory)`**은 컴퓨터 과학에서 계산 이론의 분야로, 계산 문제를 푸는 알고리즘을 복잡도에 따라 분류하여 문제의 모임을 구성하는 방법을 연구한다. 이 때 알고리즘의 수행은 실제 컴퓨터가 할 수 있지만, 평가하는 데에는 튜링 기계와 관련이 있는 정량화된 방법을 사용한다. (wikipedia)<p>
- 복잡도의 기준은 알고리즘이 소모하는 소요 시간과 메모리 사용량 등의 자원이다. 전자를 **`시간 복잡도`**, 후자를 **`공간 복잡도`**라 한다. 일반적으로 이와 같은 시공간 등의 자원은 입력의 크기에 의존하는 것으로 취급한다.<p>

- v = [3, 23, 55, 65, 76, 56, 43, 97, 34, 13]이 주어졌을 때
- 첫 번째 값을 기준으로 순차적으로 비교하며 큰 값을 찾음

```python
v = [3, 23, 55, 65, 76, 56, 43, 97, 34, 13]

def find_max(x):
    n = len(x)
    max_v = x[0]
    for i in range(1,n):
        if x[i] > max_v:
            max_v = x[i]
    return max_v
    
find_max(v)
```

## 2. 데이터 분석 기초

### 1) 확률과 통계

- **`확률(probability)`**이란?<p>
   - 확률이란 바로 **어떤 사건(event)이 일어날 가능성을 0과 1사이의 실수로 표시**하는 것<p>
   - 한 사건이 발생할 가능성이 높으면 확률은 1에 가깝고, 반대로 일어날 가능성이 낮으면 0에 가까워진다<p>
   <br>
   <br>
- **확률의 고전적 정의**<p>
   - **표본공간(sample space)**의 모든 원소가 일어날 가능성이 같다고 가정하고,
    <br>
    - 사건 A가 발생할 확률은 이산형 표본공간의 경우에<p>
    
       $$P(A) = \frac{사건\, A에\, 속하는\, 원소수(A)}{표본공간의\, 전체\, 원소수(S)}$$<p>
    <br>
    - 연속형 표본공간의 경우에<p>
    
       $$P(A) = \frac{사건\, A에\, 속하는\, 원소에\, 대한\, 측도(A)}{표본공간의\, 전체\, 원소에\, 대한\, 측도(S)}$$<p>
    <br>
    로 정의한다. 여기서 측도란 길이, 면적, 부피 등을 뜻한다<p>
    <br>
- **확률의 상대도수적 정의**<p>
   - 사건 A가 발생활 확률($P(A)$)은 같은 조건하에서 **통계적 실험**을 수없이 반복 시행했을 때<br>사건 A가 발생하는 비율, 즉 **상대도수**이다<p>
    <br>
- **확률의 공리적정의**<p>
   1. 표본공간 S에서 임의의 사건 A에 대하여, $0 \leq P(A) \leq 1$
   2. $P(S)=1$
   3. 서로 배반인 사건 $A_{1},\, A_{2}, \cdots$에 대하여<p>
    $$P(A_{1} \cup A_{2} \cup \cdots)=P(A_{1})+P(A_{2})+\cdots$$<p>
       를 만족할 때, $P(A)$를 사건 A의 확률이라 한다<p>

- **확률의 덧셈법칙**<p>
    $$P(A \cup B) = P(A) + P(B) - P(A \cap B)$$ <p>
    만일 $A \cap B=\emptyset$ 이면 $P(A \cup B)=P(A)+P(B)$이고, 사건 A, B를 서로 **배반사건(mutually exclusive events)**이라고 한다<p>
    <br>
- **확률의 곱셈법칙**
    $P(A)>0,\ P(B)>0$ 이면,<p>
    $$P(A \cap B)=P(A)P(B|A)=P(B)P(A|B)$$<p>
    이다. 만일 $P(B|A)=P(B)$ 이면 사건 A와 B를 서로 **독립사건(independent event)**이라고 한다.<p>
    이때는<p>
    $$P(A \cap B)=P(A)P(B)$$
    로 쓸 수 있다.<p>
    <br>
    <br>
- **조건부 확률(conditional probability)**<p>
    $$P(A|B)=\frac{P(A \cap B)}{P(B)}\ \ \ \ \ \ \ \ \ \ \ \ (단, P(B)>0)$$

- **`통계학(statistics)`**이란?<p>
    - 관심대상에 대하여 관련된 데이터를 수집하고 그 데이터를 요약 $\cdot$ 정리하여 이로부터 불확실한 사실에 대한 결론이나 일반적인 규칙성을 이끌어내는 방법을 연구하는 학문<p>
    <br>
- **통계학의 역할**<p>
    1. `데이터 수집` : 일반적으로 조사, 실험, 관찰 등을 통해 수집
    2. `데이터 요약` : 어떤 현상에 대해 조사 및 측정된 데이터가 갖고 있는 특징을 요약적으로 정리
    3. `데이터로부터 결론을 유도` : 이부 데이터 분석결과를 통해 일반적인 결론이나 이론을 증명<p>
    <br>
- **통계학의 구분**<p>
    - `기술통계`<p>
        - 대표되는 몇 개의 숫자로 데이터 전체를 요약하여 양적으로 축소하거나 정리된 표 그리고 그래프를 통하여 데이터의 특징을 파악<p>
        - 기술통계는 그래프에 의한 데이터의 표현과 수치에 의한 데이터의 요약과 정리<p>
        - 즉, 데이터를 대표하는 수치나 표, 그래프 등으로 요약하여 전체 특징을 파악하는 통계적 방법<p>
        <br>
    - `추측통계`<p>
        - 표본 데이터의 일부를 분석함으로써 전체에 관해 추측하고 일반화시키는 통계적 방법<p>
        - 추측통계학은 확률을 기본 개념으로 하여 불확실성의 정도를 나타냄<p>
        - 즉, 일부를 분석하여 관심대상 전체에 관해 추측하고 일반화시키는 통계적 방법<p>

- **데이터 수치 요약**<p>
    - **평균(mean)**<p>
        - 데이터의 총합을 표본의 크기로 나눈 값<p>
            <br>
        $$\bar{x}=\frac{\sum_{i=1}^{n}x_{i}}{n}$$<p>
        <br>
    - **중앙값(median)**<p>
        - n개의 관찰값을 이등분해 주는 값으로 전체 데이터가 중앙값보다 큰 값이 절반이고, 이보다 작은 값이 절반이 되는 값이다<p>
            <br>
            * n = 홀수일 때,<p>
        $$\widetilde{x}=x_{\big( \frac {n+1}{2}\big )}$$<p>
            * n = 짝수일 때,<p>
        $$\frac {x_{\big ( \frac{n}{2} \big )}+x_{\big ( \frac{n}{2}+1 \big )}}{2}$$<p>
                <br>
            여기서, $x_{k}$는 순서대로 나열했을 때, $k$ 번째 관찰값이다.<p>
            <br>
    - **최빈값(mode)**<p>
        - 데이터 중에서 발생빈도가 가장 높은 값<p>
            <br>
    - **사분위수(interquartile)**<p>
        - 순서대로 나열된 데이터 전체를 4등분하는 기준값<p>
            <br>
            * $Q_{1}=25\%,\, Q_{2}=50\%,\, Q_{3}=75\%$<p>
            * 사분위수범위(interquartile range: IQR) : $Q_{3}-Q_{1}$<p>
            * 상자그림(boxplot)을 통해 특이점을 찾아낼 수 있음<p>
            $$Q_{1}-1.5 \times IQR,\, Q_{3}+1.5 \times IQR$$<p>
            <br>
    - **분산(variance)**과 **표준편차(standard deviation)**<p>
        - 평균이 데이터의 중심위치를 나타내는 통계량으로 쓰일 때 산포의 측도로 사용된다<p>
        - 분산과 표준편차는 데이터가 평균을 중심으로 얼마나 광범위하게 분포하고 있는지는 나타내 준다<p>
            <br>
            $$\sigma^{2}=\frac {\sum_{i=1}^{N}(x_{i}-\mu)^{2}}{N}$$<p>
            $$\sigma = \sqrt{\sigma^{2}}$$<p>
            <br>
    - **변이계수(coefficient of variance)**<p>
        <br>
        - 두 데이터 간 측정단위가 다르거나 데이터값의 차이가 크게 나타나는 경우, 분산 또는 표준편자로 비교하는 거은 무의미하다<p>
        - 이런한 경우에 사용할 수 있는 측도가 변이계수이다<p>
            <br>
            $$C=\frac {\sigma}{\mu} \times 100(\%)$$

### 2) 파이썬 데이터 분석 패키지 소개

- 파이썬에서 데이터 분석에 활용하는 [numpy](https://numpy.org/), [pandas](https://pandas.pydata.org/), [matplotlib](https://matplotlib.org/) 패키지 소개<p>
    - **numpy**<p>
        - 넘파이는 수치 데이터를 다루는 패키지<p>
        - 다차원 행렬 자료구조인 **ndarray**를 통해 벡터(vector) 및 행렬(matrix)을 사용하는 선형 대수 계산에서 주로 사용<p>
        - numpy의 주요 모듈(module)은 아래와 같다<p>
            1. array() : 리스트, 튜플, 배열로부터 ndarray를 생성
            2. asarray() : 기존의 array로 부터 ndarray를 생성
            3. arange() : range와 비슷
            4. linspace(start, end, num) : [start, end] 균일한 간격으로 num개 생성
            5. logspace(start, end, num) # [start, end] log scale 간격으로 num개 생성<p><br>
    - **pandas**<p>
        - 데이터 처리를 위한 라이브러리<p>
        - 파이썬을 활용하여 데이터 분석할 때에 필수적인 라이브러리<p>
        - pandas는 CSV, text, excel, html, SQL, JSON 등 다양한 외부 데이터 파일을 읽고 데이터 프레임을 생성할 수 있다<p><br>
        - pandas의 주요 모듈은 아래와 같다<P>
            1. series
            2. dataframe
            3. panel<p><br>
    - **matplotlib**<p>
        - 데이터를 차트(chart)나 플롯(plot)으로 시각화(visualization)하는 패키지<p>
        - 데이터 분석 결과를 그래프로 표현하는 것뿐만 아니라 데이터 분석 이전에 데이터에 대한 이해를 돕기 위해 활용되기도 한다<p><br>
