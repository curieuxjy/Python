```python
import pandas as pd
```

```python
from sklearn.datasets import load_wine

data = load_wine()
df_X = pd.DataFrame(data.data, columns = data.feature_names)
```

```python
df_y = pd.DataFrame(data.target, columns=['kind(target)'])
# df['column_name'] = data_array
```

```python
data.target
```

```python
df_y.head()
```

```python
df = pd.concat([df_X, df_y], axis=1) #default axis=0, 행으로 합침
```

```python
df
```

```python
import matplotlib.pyplot as plt

plt.hist(df.loc[:, 'alcohol']) #default bins =10
```

```python
plt.boxplot(df.loc[:, 'alcohol'])
```
