```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = np.linspace(-5,5,101)
y1 = np.sin(x1)
plt.plot(x1,y1)
```

```python
# fig = 그래프 전체 영역인 캔버스를 생성하는 Figure Class 객체
# ax = 좌표평면을 생성하는 Axes Class의 객체
fig, ax = plt.subplots()

ax.set_title('Sin')
ax.set_xlabel('rad')
ax.plot(x1, y1)

handles, labels = ax.get_legend_handles_labels() #설정한 레이블을 불러오는 함수
ax.legend(handles, labels) #label 배치
# handles, labels -> Legend 클래스 객체
plt.show() # %matplotlib inline이 있다면 필요없음
```

```python
print(type(handles), handles) #empty
print(type(labels), labels) #empty
```
